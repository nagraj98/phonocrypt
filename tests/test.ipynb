### Setup

```python
from core.phonetic_engine import (
    PhoneticEngine,
    PhoneticConfig,
    Script
)

british_engine = PhoneticEngine(PhoneticConfig(language='en-gb'))
american_engine = PhoneticEngine(PhoneticConfig(language='en-us'))

sample_text = "The Quick brown fox jumps over the lazy dog"
```

#### Convert to IPA

```python
american_engine.text_to_ipa(sample_text)
```

#### Convert English to Devanagari

```python
american_engine.english_to_devanagari(sample_text, script=Script.SANSKRIT)
```

### Mixed language/script sentences

```python
sentence = [
    ('marathi', 'ते'),
    ('marathi', 'एक'),
    ('english', 'fusion'),
    ('english', 'song'),
    ('marathi', 'आहे'),
]

result = [word for lang, word in sentence]
result = ' '.join(result)
print("Mixed script - ", result)

converted_result = [american_engine.english_to_devanagari(word) if lang == 'english' else word for lang, word in sentence]
converted_result = ' '.join(converted_result)
print("Converted script - ", converted_result)  
```
